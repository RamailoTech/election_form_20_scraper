```python
import json

# Load the JSON data from a file
with open('result_json_AC_001.json') as f:
    data = json.load(f)

# Specify the desired column header to match, normalized
desired_header = "specificheadername"  # Example: the header you are looking for, in lowercase without spaces

for table in data['tables']:
    col_headers = []
    content_cells = []
    cells = table.get('cells')
    header_indices = []  # This will store the indices of columns with the desired header

    # Collect all column headers and determine indices for the desired header
    for i in range(table.get('column_count')):
        col_cells = [cell for cell in cells if cell.get('column_index') == i and cell.get('kind') == 'columnHeader']
        max_row_in_column = max([cell.get('row_index') for cell in col_cells], default=-1)

        for cell in col_cells:
            if cell.get('row_index') == max_row_in_column:
                normalized_header = cell.get('content').lower().strip()
                col_headers.append(normalized_header)  # Store normalized header
                # Check if this normalized column header is the desired one
                if normalized_header == desired_header:
                    header_indices.append(i)

    # Filter content cells based on header indices
    for cell in cells:
        if cell.get('kind') == 'content' and cell.get('column_index') in header_indices:
            content_cell_info = {
                "row_index": cell.get('row_index'),
                "column_index": cell.get('column_index'),
                "content": cell.get('content')
            }
            content_cells.append(content_cell_info)

    # Output the results for content cells matching the desired header
    print("Column Headers:")
    print(col_headers)
    print("\nContent Cells for Header '{}':".format(desired_header))
    print(content_cells)
```

```python
import json
import re

def normalize_header(header):
    """Normalize the header by removing spaces, special characters, and making it lowercase."""
    return re.sub(r'\s+|\W+', '', header).lower()

# Load the JSON data from a file
with open('result_json_AC_001.json') as f:
    data = json.load(f)

# Specify the desired column header to match, normalized
desired_header = normalize_header("Serial No. of Polling Station")  # Normalized

for table in data['tables']:
    col_headers = []
    content_cells = []
    cells = table.get('cells')
    header_indices = []  # This will store the indices of columns with the desired header

    # Collect all column headers and determine indices for the desired header
    for i in range(table.get('column_count')):
        col_cells = [cell for cell in cells if cell.get('column_index') == i and cell.get('kind') == 'columnHeader']
        max_row_in_column = max([cell.get('row_index') for cell in col_cells], default=-1)

        for cell in col_cells:
            if cell.get('row_index') == max_row_in_column:
                normalized_header = normalize_header(cell.get('content'))
                col_headers.append(normalized_header)  # Store normalized header
                # Check if this normalized column header is the desired one
                if normalized_header == desired_header:
                    header_indices.append(i)

    # Filter content cells based on header indices
    for cell in cells:
        if cell.get('kind') == 'content' and cell.get('column_index') in header_indices:
            content_cell_info = {
                "row_index": cell.get('row_index'),
                "column_index": cell.get('column_index'),
                "content": cell.get('content')
            }
            content_cells.append(content_cell_info)

    # Output the results for content cells matching the desired header
    print("Column Headers:")
    print(col_headers)
    print("\nContent Cells for Header '{}':".format(desired_header))
    print(content_cells)
```
