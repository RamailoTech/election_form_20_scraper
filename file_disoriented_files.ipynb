```python
import pandas as pd 

file_path = 'data/Maharashtra_orientation_issues.csv'

df = pd.read_csv(file_path)

l = df['AC'].tolist()

# convert it into 3 digit format and in string 

# formatted_list = ['AC_' + str(i).zfill(3) + '.pdf' for i in l ]
formatted_list = [str(i) + '.xlsx' for i in l ]

    
```

```python
print(formatted_list)
```

```python
# loop through the files in Assembly Election 2019 folder and find the filenames like AC003.pdf and copy it into another folder. 

import os 
import shutil

# source_folder = 'data/PDFs/Maharastra/Assembly Election 2019'
# destination_folder = 'data/PDFs/Maharastra/Assembly Election 2019 disoriented'
source_folder = 'results/intermediate_tables/MH/2019_GA'
destination_folder = 'results/intermediate_tables/MH/2019_GA_valid'

for file in os.listdir(source_folder):
    if file in formatted_list:
        shutil.copy(os.path.join(source_folder, file), destination_folder)
        print(f'Copied {file} to {destination_folder}')
    else:
        print(f'File {file} not found in the list')
```
